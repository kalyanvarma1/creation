```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import AutoTokenizer, AutoModel
import shutil
```

```python
import pandas as pd

# Upload the CSV to Colab first
from google.colab import files
uploaded = files.upload()

# Load dataset
df = pd.read_csv("Dataset1 - datasetforlegalai.csv")

df.head()
```

```python
train_df = df.drop(['IPC - 324'], axis=1)
target_list = [col for col in train_df.columns if col.startswith('IPC')]
print(target_list)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Get label columns
label_columns = [col for col in df.columns if col != "Narrative"]
df["labels"] = df[label_columns].values.tolist()

# Final dataset format
texts = df["Narrative"].tolist()
labels = df["labels"].tolist()
```

```python
from transformers import AutoTokenizer

model_name = "law-ai/InLegalBERT"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize
encodings = tokenizer(texts, truncation=True, padding=True, max_length=512)
```

```python
class CustomDataset(Dataset):
    def __init__(self, df, tokenizer, max_len):
        self.tokenizer = tokenizer
        self.texts = df['Narrative']
        self.targets = df[target_list].values
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, index):
        text = str(self.texts[index])
        inputs = self.tokenizer.encode_plus(
            text,
            None,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        return {
            'input_ids': inputs['input_ids'].flatten(),
            'attention_mask': inputs['attention_mask'].flatten(),
            'targets': torch.FloatTensor(self.targets[index])
        }
```

```python
# ✅ Data Split
train_size = 0.8
train_df_1 = train_df.sample(frac=train_size, random_state=200).reset_index(drop=True)
val_df = train_df.drop(train_df_1.index).reset_index(drop=True)
```

```python
MAX_LEN = 256
TRAIN_BATCH_SIZE = 16
VALID_BATCH_SIZE = 16
EPOCHS = 16
LEARNING_RATE = 2e-5
```

```python
train_dataset = CustomDataset(train_df_1, tokenizer, MAX_LEN)
val_dataset = CustomDataset(val_df, tokenizer, MAX_LEN)
```

```python
train_loader = DataLoader(train_dataset, batch_size=TRAIN_BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=VALID_BATCH_SIZE, shuffle=False)
```

```python
class InLegalBERTClassifier(nn.Module):
    def __init__(self):
        super(InLegalBERTClassifier, self).__init__()
        self.bert = AutoModel.from_pretrained('law-ai/InLegalBERT')
        self.dropout = nn.Dropout(0.3)
        self.linear = nn.Linear(768, len(target_list))

    def forward(self, input_ids, attention_mask):
        output = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        pooled_output = output.pooler_output
        x = self.dropout(pooled_output)
        return self.linear(x)
```

```python
# ✅ Initialize Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = InLegalBERTClassifier().to(device)
optimizer = torch.optim.Adam(params=model.parameters(), lr=LEARNING_RATE)
```

```python
# ✅ Optional pos_weight for class imbalance
label_freq = train_df_1[target_list].sum().values
pos_weight = torch.tensor((len(train_df_1) - label_freq) / (label_freq + 1e-5), dtype=torch.float).to(device)
loss_fn = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
```

```python
def train_model(n_epochs, train_loader, val_loader, model, optimizer, loss_fn):
    best_val_loss = float('inf')
    for epoch in range(1, n_epochs + 1):
        model.train()
        train_loss = 0
        for data in train_loader:
            input_ids = data['input_ids'].to(device)
            attention_mask = data['attention_mask'].to(device)
            targets = data['targets'].to(device)

            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask)
            loss = loss_fn(outputs, targets)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for data in val_loader:
                input_ids = data['input_ids'].to(device)
                attention_mask = data['attention_mask'].to(device)
                targets = data['targets'].to(device)

                outputs = model(input_ids, attention_mask)
                loss = loss_fn(outputs, targets)
                val_loss += loss.item()

        avg_train_loss = train_loss / len(train_loader)
        avg_val_loss = val_loss / len(val_loader)

        print(f"Epoch {epoch}: Train Loss: {avg_train_loss:.4f}, Validation Loss: {avg_val_loss:.4f}")

        # Save the best model
        if avg_val_loss < best_val_loss:
            print(f"Validation loss decreased. Saving model...")
            torch.save(model.state_dict(), '/content/drive/MyDrive/InLegalBERT_IPC_best_model.pt')
            best_val_loss = avg_val_loss

    print("Training Complete")
```

```python
train_model(EPOCHS, train_loader, val_loader, model, optimizer, loss_fn)
```

```python
def predict_top_k_ipc_sections(text, model, tokenizer, device, top_k=5):
    model.eval()
    inputs = tokenizer.encode_plus(
        text,
        None,
        add_special_tokens=True,
        max_length=256,
        padding='max_length',
        truncation=True,
        return_attention_mask=True,
        return_tensors='pt'
    )

    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)

    with torch.no_grad():
        logits = model(input_ids, attention_mask)
        probabilities = torch.sigmoid(logits).cpu().numpy()[0]

    # Top-K IPCs
    top_indices = probabilities.argsort()[-top_k:][::-1]
    top_ipcs = [(target_list[i], probabilities[i]) for i in top_indices]
    return top_ipcs
```

```python
import torch
class IPCDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        return {
            **{k: torch.tensor(v[idx]) for k, v in self.encodings.items()},
            "labels": torch.tensor(self.labels[idx], dtype=torch.float)
        }

    def __len__(self):
        return len(self.labels)

dataset = IPCDataset(encodings, labels)
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    problem_type="multi_label_classification",
    num_labels=len(label_columns)
)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=5,
    per_device_train_batch_size=4,
    logging_dir="./logs",
    logging_steps=10
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)
```

```python
trainer.train()
```
